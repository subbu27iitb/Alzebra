```python
Gx=np.array([[1,0,-1],[2,0,-2],[1,0,-1]])
Gy= np.array([[1,2,1],[0,0,0],[-1,-2,-1]])

nx,ny,nz = input_image.shape
sobel_matrix1=np.zeros(shape=(nx,ny))
sobel_matrix2=np.zeros(shape=(nx,ny))
sobel_matrix3=np.zeros(shape=(nx,ny))
gray_image1=input_image[:,:,0]
for i in range(nx-2):
  for j in range(ny-2):
    gx=np.sum(np.multiply(gray_image1[i:i+3,j:j+3],Gx))
    gy=np.sum(np.multiply(gray_image1[i:i+3,j:j+3],Gy))
    r=np.sqrt(gx**2+gy**2)
    sobel_matrix1[i+1,j+1]=r

gray_image2=input_image[:,:,1]
for i in range(nx-2):
  for j in range(ny-2):
    gx=np.sum(np.multiply(gray_image2[i:i+3,j:j+3],Gx))
    gy=np.sum(np.multiply(gray_image2[i:i+3,j:j+3],Gy))
    r=np.sqrt(gx**2+gy**2)
    sobel_matrix2[i+1,j+1]=r

gray_image3=input_image[:,:,2]
for i in range(nx-2):
  for j in range(ny-2):
    gx=np.sum(np.multiply(gray_image3[i:i+3,j:j+3],Gx))
    gy=np.sum(np.multiply(gray_image3[i:i+3,j:j+3],Gy))
    r=np.sqrt(gx**2+gy**2)
    sobel_matrix3[i+1,j+1]=r


sobel_matrix=sobel_matrix1+sobel_matrix2+sobel_matrix3

#sobel_matrix
plt.imshow(sobel_matrix, cmap='gray')
print(sobel_matrix.shape)


```

#**Find polynomial equation coefiicents using Gradient descent algorithm**

```python
import numpy as np
import torch

poly=np.poly1d([1,2,3]) #Actual coefficents
print(poly)
print(poly(1))
print("***********************************")
N=20 #no.of samples
# Generate training data roughly between -6 to +6
X=np.random.randn(N,1)*3
Y=poly(X)

##start of optimization coding sample

# Prepare input as an array of shape (N,3)
XX=np.hstack([X*X,X,np.ones_like(X)]) #[X^2, X, 1]
# Prepare tensors
w=torch.randn(3,1,requires_grad=True) #3 coefficents as a, b & c in  ax^2+bx+c
x=torch.tensor(XX, dtype=torch.float32) #input in training data
y=torch.tensor(Y,dtype=torch.float32) #target in training data
optimizer = torch.optim.NAdam([w],lr=0.01)
print(w)
print("*********************************************")

for _ in range(1000):
  optimizer.zero_grad()
  y_pred=x@w
  mse = torch.mean(torch.square(y - y_pred))
  mse.backward()
  optimizer.step()

print(w)
```

```python
import torch
import random
import numpy as np

A=torch.tensor(random.random(), requires_grad=True)
B=torch.tensor(random.random(), requires_grad=True)
C=torch.tensor(random.random(), requires_grad=True)
D=torch.tensor(random.random(), requires_grad=True)

itr=100 #no. of iterations
optimizer=torch.optim.NAdam([A,B,C,D], lr=0.01)

for _ in range(itr):
  y1=A+B-9
  y2=C-D-1
  y3=A+C-8
  y4=B-D-2

  squerr = y1*y1 + y2*y2 + y3*y3 + y4*y4
  optimizer.zero_grad()
  squerr.backward()
  optimizer.step()

print(A)
print(B)
print(C)
print(D)
```

#**Moving Average**

```python
import numpy as np
def moving_average(data, window_size):
    moving_average = []
    for i in range(len(data) - window_size + 1):
        moving_average.append(sum(data[i:i + window_size]) / window_size)
    return moving_average


data= np.arange(1,20+1,1)
print()
moving_average(data, 3)
```

#**Exponential Moving Average**

```python
def exponential_moving_average(data, window_size):
    alpha = 2 / (window_size + 1)
    exponential_moving_average = [data[0]]
    for i in range(1, len(data)):
        weighted_sum = alpha * data[i] + (1 - alpha) * exponential_moving_average[-1]
        exponential_moving_average.append(weighted_sum / window_size)
    return exponential_moving_average



data= np.arange(1,20+1,1)
window_size=3
#alpha=0.395
exponential_moving_average(data, window_size)
```

```python
import math

def f(x):
    return x**3 - 3*x**2 + 4

def df(x):
    return 3*x**2 - 6*x # derivative of f(x)

def bisection(f, a, b, tol):
    # check if f(a) and f(b) have opposite signs
    if f(a) * f(b) > 0:
        print("No root in this interval")
        return None
    # iterate until the interval is smaller than the tolerance
    while abs(b - a) > tol:
        # find the midpoint of the interval
        c = (a + b) / 2
        # check if f(c) is zero
        if f(c) == 0:
            return c
        # check if f(c) has the same sign as f(a)
        elif f(c) * f(a) > 0:
            # update the lower bound
            a = c
        else:
            # update the upper bound
            b = c
    # return the midpoint as the root
    return c

# find the root of df(x) between -1 and 1 with tolerance 0.001
x_min = bisection(df, -1, 1, 0.001)
print(x_min) # this is the value of x that minimizes f(x)
print(f(x_min)) # this is the minimum value of f(x)
```
